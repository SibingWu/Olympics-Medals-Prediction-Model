### This notebook file is used to generate the Dataset for 2020 Olympic

since the 2020 Olympic statistic values are not given, we have the predict the needed value with the previous year data

```python
# Import required libraries and read test and train data set
import pandas as pd
import numpy as np
```

```python
# import the Summer Dataset with the given data as a Help Dataset
help_df = pd.read_csv("Generated Dataset/summer_dataset.csv")
help_df = help_df.drop(["Unnamed: 0"], axis = 1)
help_df.head()
```

```python
# import the Winter Dataset with the given data as a Help Dataset
help_df_winter = pd.read_csv("Generated Dataset/winter_dataset.csv")
help_df_winter = help_df_winter.drop(["Unnamed: 0"], axis = 1)
help_df_winter.head()
```

```python
df_2020 = pd.DataFrame(columns = help_df.columns)
df_2020.columns
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# extract the data from the latest Summer Olympic
data_2016_summer = pd.DataFrame(help_df[help_df["Year"] == 2016])
print(data_2016_summer.shape)
data_2016_summer.head()
```

```python
# extract the data from the latest Winter Olympic
data_2014_winter = pd.DataFrame(help_df_winter[help_df_winter["Year"] == 2014])
print(data_2014_winter.shape)
data_2014_winter.head()
```

```python
# create the DataFrame for 2020 Olympic
df_2020_summer = pd.DataFrame(columns = help_df.columns)
```

```python
# Feed in the Data for 2020 Olympic
df_2020_summer["CountryCode"] = data_2016_summer["CountryCode"]
df_2020_summer["Year"] = data_2016_summer["Year"]
df_2020_summer["TotalParticipants"] = data_2016_summer["TotalParticipants"]
df_2020_summer["PrevTotalParticipants"] = data_2016_summer["TotalParticipants"]
df_2020_summer["PrevGoldMedals"] = data_2016_summer["GoldMedals"]
df_2020_summer["PrevSilverMedals"] = data_2016_summer["SilverMedals"]
df_2020_summer["PrevBronzeMedals"] = data_2016_summer["BronzeMedals"]
df_2020_summer["PrevTotalMedals"] = data_2016_summer["TotalMedals"]
```

```python
df_2020_summer.head()
```

```python
df_2020_summer = df_2020_summer.reset_index(drop = True)
```

```python
NOC = list(df_2020_summer["CountryCode"])
Population = list(df_2020_summer["PopulationSize"])
Year = list(df_2020_summer["Year"])
NOC_df = pd.DataFrame(NOC)
```

```python
host_city = [0] * 197
```

```python
# The host city column for 2020 Summer Olympic should be a column of 0s except Japan which is the Host Country of 2020 Olympic
host_city_df = pd.concat([NOC_df, pd.DataFrame(host_city)], axis = 1)
host_city_df.columns = ["NOC", "HostCity"]
host_city_df = host_city_df.set_index("NOC")
host_city_df.loc["JPN", "HostCity"] = 1
host_city_df = host_city_df.reset_index(drop = True)
```

```python
# Feed in the data for 2020 Olympic
df_2020_summer["HostCity"] = host_city_df["HostCity"]
```

```python
df_2020_summer.head()
```

```python
# use Linear Regression with given previous year data to predict the GDP data for each country in 2020
NOC_gdp_list = []
for i in NOC:    
    NOC_gdp = pd.DataFrame(help_df[help_df["CountryCode"] == i]["GDPPerCapita"])
    NOC_year = pd.DataFrame(help_df[help_df["CountryCode"] == i]["Year"])
    linreg = LinearRegression()
    linreg.fit(NOC_year, NOC_gdp)
    NOC_gdp_pred = linreg.predict(pd.DataFrame([2020]))
    temp = list(NOC_gdp_pred[0])
    if temp[0] < 0: # if the value is negative, then use the previous year data
        temp[0] = NOC_gdp["GDPPerCapita"].iloc[-1] 
    NOC_gdp_list.extend(temp)
```

```python
df_2020_summer["GDPPerCapita"] = NOC_gdp_list
```

```python
# use Linear Regression with given previous year data to predict the Population data for each country in 2020
NOC_population_list = []
for i in NOC:    
    NOC_population = pd.DataFrame(help_df[help_df["CountryCode"] == i]["PopulationSize"])
    NOC_year = pd.DataFrame(help_df[help_df["CountryCode"] == i]["Year"])
    linreg = LinearRegression()
    linreg.fit(NOC_year, NOC_population)
    NOC_population_pred = linreg.predict(pd.DataFrame([2020]))
    temp = list(NOC_population_pred[0])
    if temp[0] < 0: # if the value is negative, then use the previous year data
        temp[0] = NOC_population["PopulationSize"].iloc[-1]
    NOC_population_list.extend(temp)
```

```python
df_2020_summer["PopulationSize"] = NOC_population_list
```

```python
data_2014_winter = data_2014_winter.set_index("CountryCode")
df_2020_summer = df_2020_summer.set_index("CountryCode")
```

```python
# Feed in the Data for 2020 Olympic
for i in df_2020_summer.index:
    if i in data_2014_winter.index:
        df_2020_summer.loc[i, "PrevWinterTotalParticipants"] = data_2014_winter.loc[i, "TotalParticipants"]
        df_2020_summer.loc[i, "PrevWinterGoldMedals"] = data_2014_winter.loc[i, "GoldMedals"]
        df_2020_summer.loc[i, "PrevWinterSilverMedals"] = data_2014_winter.loc[i, "SilverMedals"]
        df_2020_summer.loc[i, "PrevWinterBronzeMedals"] = data_2014_winter.loc[i, "BronzeMedals"]
        df_2020_summer.loc[i, "PrevWinterTotalMedals"] = data_2014_winter.loc[i, "TotalMedals"]
```

```python
# fill all the Null value with 0
df_2020_summer["PrevWinterTotalParticipants"].fillna(value = 0, inplace = True)
df_2020_summer["PrevWinterGoldMedals"].fillna(value = 0, inplace = True)
df_2020_summer["PrevWinterSilverMedals"].fillna(value = 0, inplace = True)
df_2020_summer["PrevWinterBronzeMedals"].fillna(value = 0, inplace = True)
df_2020_summer["PrevWinterTotalMedals"].fillna(value = 0, inplace = True)
```

```python
# create the csv file
df_2020_summer.to_csv("Generated Dataset/2020 Olympics.csv") # keep the index(CountryCode) this time
```
